## Projet: Calcul des mensualités d'un pret                                               28/12/2024
Calcul des mensualités pour un emprunt E à taux b sur N mensualités

Les mensualités sont constantes et on veut calculer
la part du remboursement de l'emprunt et la part des intérets.


L'emprunt est fait à étape 0, au bout d'un mois on a:
\begin{equation}
m_1  = Eb + c_1
\end{equation}
les mois suivants on a:

\begin{eqnarray}
m_2 &=& (E-c_1)b + c_2      \\
m_3 &=& (E-c_1 - c_2)b + c_3 \\
m_N &=& (E-c_1 -c_2 -c_3 ... -c_{N-1})b + c_N 
\end{eqnarray}

on doit avoir:
\begin{equation}
\sum_{k=1}^N c_k = E
\end{equation}
et pour les intérets:
\begin{equation}
\sum_{k=1}^N I_k = \sum_{k=1}^N m_k - E
\end{equation}

 Les mensualités sont constantes, donc $m_1=m_2$, on:
\begin{equation}
Eb + c_1 = (E-c_1)b + c_2
\end{equation}
on tire:
\begin{equation}
c_2 = c_1(1+b)
\end{equation}
de meme $m_2=m_3$, on obtient:
\begin{equation}
c_3 = c_2(1+b) = c_1(1+b)^2
\end{equation}
on obtient donc pour la kieme et la Nieme etape:
\begin{eqnarray}
c_k = c_1(1+b)^{k-1} \\
c_N = c_1(1+b)^{N-1}
\end{eqnarray}
calculons la valeur de $c_1$:
\begin{eqnarray}
E &=& c_1 + c_2 + c_3 + ... c_N \\
  &=& c_1 + c_1(1+b) + c_1(1+b)^2 + ...c_1(1+b)^{N-1} \\
E(1+b)&=& c_1(1+b) + c_1(1+b)^2 + ...c_1(1+b)^{N} \\
Eb&=& c_1(1+b)^N - c_1 = c_1((1+b)^N - 1)
\end{eqnarray} 
On a donc:
\begin{equation}
\color{blue}
{c_1 = \frac{Eb}{(1+b)^N - 1}}
\end{equation}
la mensualité est donc:
\begin{eqnarray}
m_1 = Eb + c_1 = Eb\left(1 + \frac{1}{(1+b)^N - 1}\right)
\end{eqnarray}

On peut exprimer $m_1$ d'une autre facon:
\begin{eqnarray}
Eb\left(1 + \frac{1}{(1+b)^N - 1}\right)&=&Eb\left(\frac{(1+b)^N-1+1}{(1+b)^N-1}\right) \\
& =& Eb\left(\frac{(1+b)^N}{(1+b)^N-1}\right) \\
&=& Eb\left(\frac{(1+b)^N}{(1+b)^N}\right)\frac{1}{1-\frac{1}{(1+b)^N}} \\
&=& \color{blue}{\frac{Eb}{1-(1+b)^{-N}}}
\end{eqnarray}

calcul des intérets
\begin{eqnarray}
I_1 &=& Eb \\
I_2 &=& (E-c_1)b \\
I_3 &=& (E-c_1-c_2)b = (E-c_1 -c_1(1+b)^2 \\
I_k &=& (E -(c_1 + c_1(1+b) +.....c_1(1+b)^{k-2})
\end{eqnarray}

On calcul la somme des termes:
\begin{eqnarray}
A     &=& c_1 + c_1(1+b) +.....c_1(1+b)^{k-2} \\
A(1+b)&=& c_1(1+b) + ....c_1(1+b)^{k-1} \\
Ab    &=& c_1(1+b)^{k-1} -c_1 = c_1((1+b)^{k-1}-1) \\
A     &=& \frac{c_1}{b}((1+b)^{k-1}-1)
\end{eqnarray}
donc:
\begin{eqnarray}
I_k&=& (E - \frac{c_1}{b}((1+b)^{k-1}-1))b \\
   &=& Eb - c_1((1+b)^{k-1}-1) \\
   &=& Eb -\left(\frac{Eb}{(1+b)^N-1}(1+b)^{k-1} -1\right) \\
   &=& \frac{Eb}{(1+b)^N-1} \left((1+b)^N-1-\left((1+b)^{k-1} -1\right)\right) \\
   &=& \frac{Eb}{(1+b)^N-1} \left((1+b)^N-\left((1+b)^{k-1}\right)\right) \\
   &=& \color{blue}{c1\left((1+b)^N-(1+b)^{k-1}\right)}
\end{eqnarray}
verifions que :
\begin{eqnarray}
I_k + c_k &=& c_1\left((1+b)^N-(1+b)^{k-1}\right) + c_1(1+b)^{k-1} \\
          &=& c_1\left((1+b)^N\right) \\
          &=& \frac{Eb}{(1+b)^N-1}(1+b)^N \\
          &=& Eb\frac{(1+b)^N-1 + 1}{(1+b)^N-1} \\
          &=& Eb\left(1 + \frac{1}{(1+b)^N - 1}\right)\\
          &=& m_1
\end{eqnarray}
On va aussi calculer le cout du pret:
\begin{eqnarray}
cout &=& \sum_{k=1}^{N} = I_k \\
     &=& \sum_{k=1}^{N} \left (c_1\left((1+b)^N - (1+b)^{k-1}\right)\right) \\
     &=& Nc_1(1+b)^N - c_1\left(1 + (1+b)+...(1+b)^{N-1}\right) \\
     &=& c_1 N (1+b)^N - \frac{c_1}{b}\left((1+b)^{N-1}-1\right) \\
     &=& c_1 N (1+b)^N -E
\end{eqnarray}

```python
import itertools
import pandas as pd
E = 1000
b = 0.03/12
N = 1*12


def mensualite(E,b,N):
    c1 = E*b /((1+b)**N -1)
    return round(E*b + c1,2)

def cn(step,Emprunt=E, taux=b, duree=N):
    return Emprunt*taux*((1.+taux)**(step-1.))/((1+taux)**duree -1.)

def In(step,Emprunt=E, taux=b, duree=N):

    num = Emprunt * taux
    den = (1+taux)**duree - 1.
    therm1 = (1+taux)**duree
    therm2 = -(1+taux)**(step-1.)
    return (therm1+therm2) * num / den


print(f"m:{mensualite(E,b,N)}")
```

```python
def cout(E,b,N):
    c1 = E*b /((1+b)**N -1)
    return c1 * N*(1+b)**N - E


c1 = cout(E,b,N)

print(f"c1:{c1}")
```

```python
class Pret:
    """ Class qui calcul la mensualité, le cout,
    le capital rembourse et les interets à l'etape k"""
    def __init__(self,E=63000,b=0.035,N=240):
        self.E = E
        self.b = b/12
        self.N = N
        self.c1 = self.E*self.b /((1.+self.b)**self.N -1.)
        
    def mensualite(self):
        mens = self.E*self.b + self.c1
        return round(mens,2)
    
    def cout(self):
        cout = self.c1 * self.N*(1+self.b)**self.N - self.E
        return round(cout,2)
    
    def cn(self,k):
        ck = self.c1*(1+self.b)**(k-1.)
        return round(ck,2)
        
    def In(self,k):

        Ik = self.c1 * ((1.+self.b)**self.N - (1. + self.b)**(k-1))
        return round(Ik,2)
       
    def __repr__(self):
        return f"mensualite et cout:{self.mensualite()}, {self.cout()}"
    
    
   
P1 = Pret()
print(P1)
print(P1.cn(1))
print(P1.In(1))
print(P1.b)
print(P1.mensualite())
```

```python
def make_data(taux,Nmens):
    
    Emin, Emax,Estep = 10000, 80000, 10000
    
    length = len(range(Emin,Emax,Estep))
    P1.b = taux
    P1.N = Nmens
    data = {"Emprunt":[e for e in range(Emin,Emax,Estep)],
            "taux annuel":list(itertools.repeat(P1.b,length)),
            "durée":list(itertools.repeat(P1.N,length)),
            "mensualité":[Pret(e,b=taux,N=Nmens).mensualite() for e in range(Emin,Emax,Estep)],
            "cout":[Pret(e,b=taux,N=Nmens).cout() for e in range(Emin,Emax,Estep)]}
    
    df = pd.DataFrame.from_dict(data)
    
    return df

def make_dataframe():
    tx = 0.02
    periode = 60 # 5 ans
    list_taux = [tx*e for e in range(1,4,1)]
    list_mens = [periode*e for e in range(1,4,1)]
    
    df_taux = [make_data(taux=list_taux[i],Nmens=60) for i in range(0,3,1)]
    df_mens = [make_data(tx,list_mens[i]) for i in range(0,3,1)]
    
    return df_taux,df_mens
```

```python
def print_dataframe():
    e=2
    df_taux, df_mens = make_dataframe()
    for df in df_taux:
        print(50*"*")
        #print(f"essai{df["durée"].iloc[0]}")
        print(f"Pret sur {df['durée'].iloc[0]} mois à taux:{df['taux annuel'].iloc[0]} %")
        print(50*"*")
        print(df)
        print(50*"-")
    for df in df_mens:
        print(50*"*")
        print(f"Pret sur {df['durée'].iloc[0]} mois à taux:{df['taux annuel'].iloc[0]} %")
        print(50*"*")
        print(df)
        print(50*"-")
        
    return df

df = print_dataframe()
dfnew = df
dfnew["cout annualisé"] = (dfnew["cout"]/12).round(2)
dfnew
```
